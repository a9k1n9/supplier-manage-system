```python
import pandas as pd
# 读取清洗好的企业主表
df = pd.read_excel("供应商名录_正常版.xlsx", sheet_name="企业主表")
# 只看正常有效供应商，过滤作废数据
df_valid = df[df["数据状态"] == "正常"]

# 统计各行业企业数量
industry_count = df_valid["行业"].value_counts()
print("各行业供应商数量：")
print(industry_count)
```

```python
import matplotlib.pyplot as plt

# 设置中文（必须加，否则行业中文乱码）
plt.rcParams["font.family"] = ["SimHei", "Microsoft YaHei", "Times New Roman"]
plt.rcParams["axes.unicode_minus"] = False

# 画布大小
plt.figure(figsize=(12, 6))
bars = plt.bar(industry_count.index, industry_count.values, color="#2E86AB")

# 在柱子上标注数字
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f"{int(height)}", ha="center", va="bottom")

plt.title("各行业供应商数量分布", fontsize=14)
plt.xlabel("行业类别")
plt.ylabel("供应商数量")
plt.xticks(rotation=45, ha="right") # x轴文字旋转防重叠
plt.tight_layout() # 自动适配布局，防止文字截断
plt.savefig("行业数量柱状图.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(9, 9))
# 行业过多时，把数量极少的合并为“其他”
threshold = 3
main_industry = industry_count[industry_count >= threshold]
other_sum = industry_count[industry_count < threshold].sum()
if other_sum > 0:
    main_industry["其他行业"] = other_sum

plt.pie(main_industry.values, labels=main_industry.index, autopct="%1.1f%%",
        startangle=90)
plt.title("供应商行业占比饼图")
plt.savefig("行业占比饼图.png", dpi=300)
plt.show()
```

```python
# 统计每个行业：失信/未失信数量
industry_bad = pd.crosstab(df_valid["行业"], df_valid["是否失信"])
industry_bad.plot(kind="bar", stacked=True, figsize=(12,6), color=["#57CC99","#E63946"])
plt.title("各行业失信供应商分布")
plt.xlabel("行业")
plt.ylabel("企业数量")
plt.xticks(rotation=45, ha="right")
plt.legend(title="是否失信")
plt.tight_layout()
plt.savefig("行业失信堆叠图.png", dpi=300)
plt.show()
```

```python
from pyecharts.charts import Bar
from pyecharts import options as opts

# 组装数据 [(行业名,数量),...]
data_list = list(zip(industry_count.index.tolist(), industry_count.values.tolist()))

bar = (
    Bar(init_opts=opts.InitOpts(width="1200px", height="600px"))
    .add_xaxis([i[0] for i in data_list])
    .add_yaxis("供应商数量", [i[1] for i in data_list])
    .set_global_opts(
        title_opts=opts.TitleOpts(title="各行业供应商统计"),
        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=40)),
        datazoom_opts=opts.DataZoomOpts(), # 底部滑动缩放
    )
)
bar.render("行业柱状图.html") # 生成html，双击打开即可
```
